# 프로젝트 설명

Yahoo finance에서 선정한 세 가지의 주식의 종가를 시계열 데이터로 간주하고 ARIMA 모델에 적용하여 종가를 예측해보자

### 패키지 임포트

```python
import os
import math
import warnings 
import numpy as np
import pandas as pd
import pmdarima as pm
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore') #경고 무시

from pmdarima.arima import ndiffs
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

### 어제 오른 내 주식, 과연 내일은?

#### 데이터 가져오기

```python
'''
4월28일 데이터 추출
SOXL (미국 반도체 ETF x3 레버리지)
NVDA (엔베디아)
MRNA (모더나)
'''
soxl_filepath = os.getenv('HOME') + '/aiffel/stock_prediction/data/SOXL.csv'
nvda_filepath = os.getenv('HOME') + '/aiffel/stock_prediction/data/NVDA.csv'
mrna_filepath = os.getenv('HOME') + '/aiffel/stock_prediction/data/MRNA.csv'

soxl_df = pd.read_csv(soxl_filepath, index_col='Date', parse_dates=True)
nvda_df = pd.read_csv(nvda_filepath, index_col='Date', parse_dates=True)
mrna_df = pd.read_csv(mrna_filepath, index_col='Date', parse_dates=True)
soxl_ts = soxl_df['Close']
nvda_ts = nvda_df['Close']
mrna_ts = mrna_df['Close']

print(soxl_ts.head(), end = "\n\n\n")
print(nvda_ts.head(), end = "\n\n\n")
print(mrna_ts.head(), end = "\n\n\n")
```

```python
'''
원본 시계열 데이터와 함께 이동평균, 이동표준편차를 그려주는 함수
'''
def plot_rolling_statistics(timeseries, window=12):
    
    rolmean = timeseries.rolling(window=window).mean()  # 이동평균 시계열
    rolstd = timeseries.rolling(window=window).std()    # 이동표준편차 시계열

     # 원본시계열, 이동평균, 이동표준편차를 plot으로 시각화해 본다.
    orig = plt.plot(timeseries, color='blue',label='Original')    
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
```

```python
'''
Dickey-Fuller Test 결과 요약 함수
'''
def augmented_dickey_fuller_test(timeseries):
    # statsmodels 패키지에서 제공하는 adfuller 메서드를 호출합니다.
    dftest = adfuller(timeseries, autolag='AIC')  
    
    # adfuller 메서드가 리턴한 결과를 정리하여 출력합니다.
    print('Results of Dickey-Fuller Test:')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
'''
앞선 두 함수함와 함께,
시계열데이터 원본, 추세, 계절성,
그리고 원본 데이터에서 추세와 계절성을 제거한
residual 데이터를 시각화 해주는 함수
'''

def residual_vis(ts):
    ts = ts.interpolate(method='time')
    
    #if not ts[ts.isna()].empty:
    #    pass'''waring'''
    
    ts_log = np.log(ts)
    
    # 정성적 그래프 분석
    plot_rolling_statistics(ts_log, window=12)

    #정량적 Augmented Dicky-Fuller Test
    #augmented_dickey_fuller_test(ts_log)

    #시계열 분해 (Time Series Decomposition)

    decomposition = seasonal_decompose(ts_log, model='multiplicative', period = 30) 

    trend = decomposition.trend
    seasonal = decomposition.seasonal
    residual = decomposition.resid

    plt.figure(figsize=(10,5), dpi=100)
    plt.subplot(411)
    plt.plot(ts_log, label='Original')
    plt.legend(loc='best')
    plt.subplot(412)
    plt.plot(trend, label='Trend')
    plt.legend(loc='best')
    plt.subplot(413)
    plt.plot(seasonal,label='Seasonality')
    plt.legend(loc='best')
    plt.subplot(414)
    plt.plot(residual, label='Residuals')
    plt.legend(loc='best')
    plt.tight_layout()
    
    residual.dropna(inplace=True)
    augmented_dickey_fuller_test(residual)
    
    return ts_log    
```

## SOXL 데이터

```python
ts_log = residual_vis(soxl_ts)
```

```python
'''
시계열 데이터를
80:20 비율로
훈련 데이터 세트와 테스트 데이터 세트를 나움
'''
train_data, test_data = ts_log[:int(len(ts_log)*0.8)], ts_log[int(len(ts_log)*0.8):]
plt.figure(figsize=(10,6))
plt.grid(True)
plt.plot(ts_log, c='r', label='training dataset')  # train_data를 적용하면 그래프가 끊어져 보이므로 자연스러운 연출을 위해 ts_log를 선택
plt.plot(test_data, c='b', label='test dataset')
plt.legend()
plt.show()
```

```python
'''
원본 시계열 데이터로 dickey_fuller_test 수행한 결과와
ACF, PACF 시각화 제시
'''
augmented_dickey_fuller_test(ts_log)

plt.figure(figsize=(10,5), dpi=100)
plot_acf(ts_log,)   # ACF : Autocorrelation 그래프 그리기
plot_pacf(ts_log)  # PACF : Partial Autocorrelation 그래프 그리기
plt.show()
```

```python
'''
1차 차분한 데이터로 dickey_fuller_test 수행한 결과와
ACF, PACF 시각화 제시
'''
# 1차 차분 구하기
diff_1 = ts_log.diff(periods=1).iloc[1:]
diff_1.plot(title='Difference 1st')

augmented_dickey_fuller_test(diff_1)

plt.figure(figsize=(10,5), dpi=100)
plot_acf(diff_1)   # ACF : Autocorrelation 그래프 그리기
plot_pacf(diff_1)  # PACF : Partial Autocorrelation 그래프 그리기
plt.show()
```

```python
'''
2차 차분한 데이터로 dickey_fuller_test 수행한 결과와
ACF, PACF 시각화 제시
'''
# 혹시 필요한 경우 2차 차분 구하기
diff_2 = diff_1.diff(periods=1).iloc[1:]
diff_2.plot(title='Difference 2nd')

augmented_dickey_fuller_test(diff_2)

# ACF, PACF 그려보기 -> p,q 구하기

plt.figure(figsize=(10,5), dpi=100)
plot_acf(diff_2,)   # ACF : Autocorrelation 그래프 그리기
plot_pacf(diff_2)  # PACF : Partial Autocorrelation 그래프 그리기
plt.show()
```

```python
'''
2차 차분에서 시계열 데이터를 시각화 해보았을때, stationality를 확인할 수 있었고,
PACF에서 절단점이 25이기 때문에 p = 25,
ACF에서 절단점이 7이어서, q= 7로 설정했다.
'''
# Build Model
model = ARIMA(train_data, order=(25, 2, 7))  
fitted_m = model.fit() 

print(fitted_m.summary())
```

```python
# Forecast : 결과가 fc에 담깁니다. 
fc = fitted_m.forecast(len(test_data), alpha = 0.05)  # 95% conf
fc = np.array(fc)
# Make as pandas series
fc_series = pd.Series(fc, index=test_data.index)   # 예측결과

# Plot
#plt.figure(figsize=(10,5), dpi=100)
plt.plot(train_data, label='training')
plt.plot(test_data, c='b', label='actual price')
plt.plot(fc_series, c='r',label='predicted price')
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(np.exp(test_data), np.exp(fc))
print('MSE: ', mse)

mae = mean_absolute_error(np.exp(test_data), np.exp(fc))
print('MAE: ', mae)

rmse = math.sqrt(mean_squared_error(np.exp(test_data), np.exp(fc)))
print('RMSE: ', rmse)

mape = np.mean(np.abs(np.exp(fc) - np.exp(test_data))/np.abs(np.exp(test_data)))
print('MAPE: {:.2f}%'.format(mape*100))
```

# 회고

시계열 데이터에서 이전의 패턴을 가지고 이후에 패턴을 분석하는 것과, 차분함으로써 분산과 평균이 일정한 정적임을 찾는 과정에서 수학의 아름다움을 느낄 수 있었다.
데이터는 SOXL, NVDA, MRNA의 세개의 데이터를 사용했다. 원본 시계열 데이터와 함께 이동평균, 이동표준편차를 시각화하는 함수를 작성했고, Dickey-Fuller Test 결과를 요약 해주는 함수를 작성했다. 앞선 두 함수함와 함께 시계열데이터 원본, 추세, 계절성, 그리고 원본 데이터에서 추세와 계절성을 제거한 residual 데이터를 시각화 해주는 함수를 작성했다. 

시계열 데이터는 80:20 비율로 훈련 데이터 세트와 테스트 데이터 세트를 나누었다. 원본 시계열 데이터를 1차 차분, 2차 차분을 진행하고 Dickey-Fuller Test를 수행하여 stationarity에 대한 통계적 유의성을 확보했다. ACF, PACF 시각화를 진행하였고 이를 기반으로 데이터별 p, d, q 값을 설정했던 기준은 아래와 같다.

SOXL 데이터
2차 차분한 데이터에서 stationality를 확인할 수 있었고, PACF에서 8에서 최초로 절단점이 발견됬지만, 그 이후로 절단점을 벗어나는 값이 확인됬기 때문에 최댓값이 25로 p를 설정했고, ACF에서 3에서 절단점이 최초로 발견됬지만, 그 이후로 절단점을 벗어나지 않는 값이 10이였지만, MAPE값이 15퍼 이상의 값이 도출되어서 점차 내리다가 7이라는 최적의 q 값을 찾아 (p,d,q) = (25, 2, 7)로 설정했다.

NVDA 데이터
2차 차분한 데이터에서 stationality를 확인할 수 있었고, PACF에서 7에서 최초로 절단점이 발견됬지만, 그 이후로 절단점을 벗어나는 값이 확인됬기 때문에 최댓값이 25로 p를 설정했고, ACF에서 2에서 절단점이 최초로 발견되었고, 그 이후로 절단점을 벗어나지 않았기 때문에 2이라는 최적의 q 값을 찾아 (p,d,q) = (25, 2, 2)로 설정했다.

MRNA 데이터
1차 차분한 데이터에서 stationality를 확인할 수 있었고, PACF에서 1에서 최초로 절단점이 발견됬지만, 그 이후로 절단점을 벗어나는 값이 확인됬기 때문에 최댓값이 25로 p를 설정했고, ACF에서 1에서 절단점이 최초로 발견되었고, 그 이후로 절단점을 벗어났고 그 다음으로 발견된 안정적인 절단점이 6이기 때문에 이라는 최적의 q 값을 찾아 (p,d,q) = (25, 1, 6)으로 설정했다.

MAPE 결과 값은 SOXL, NVDA, MRNA데이터 순으로 10.47, 9.89, 13.69로 루브릭에서 요구하는 15%이하의 값이 도출되었다.
